```python
# Gerekli ktphaneleri import etme
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Veriyi okuma
df = pd.read_csv("./diabetes.csv")
y = df["Outcome"]
X = df.drop(['Outcome'], axis=1)

# Veriyi eitim ve test olarak ayrma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# Veriyi lekleme
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modelleri oluturma ve eitim
knn_tuned = KNeighborsClassifier().fit(X_train_scaled, y_train)
loj_model = LogisticRegression(max_iter=1000).fit(X_train_scaled, y_train)
svm_tuned = SVC().fit(X_train_scaled, y_train)
mlpc_tuned = MLPClassifier().fit(X_train_scaled, y_train)
cart_tuned = DecisionTreeClassifier().fit(X_train_scaled, y_train)
rf_tuned = RandomForestClassifier().fit(X_train_scaled, y_train)
gbm_tuned = GradientBoostingClassifier().fit(X_train_scaled, y_train)
catb_tuned = CatBoostClassifier().fit(X_train_scaled, y_train)
lgbm_tuned = LGBMClassifier().fit(X_train_scaled, y_train)
xgb_tuned = XGBClassifier().fit(X_train_scaled, y_train)

# Model isimleri ve doruluk oranlarn saklamak iin bir liste oluturma
modeller = [
    knn_tuned,
    loj_model,
    svm_tuned,
    mlpc_tuned,
    cart_tuned,
    rf_tuned,
    gbm_tuned,
    catb_tuned,
    lgbm_tuned,
    xgb_tuned
]

sonuc = []
sonuclar = pd.DataFrame(columns=["Modeller", "Accuracy"])

for model in modeller:
    isimler = model.__class__.__name__
    y_pred = model.predict(X_test_scaled)
    dogruluk = accuracy_score(y_test, y_pred)
    sonuc = pd.DataFrame([[isimler, dogruluk * 100]], columns=["Modeller", "Accuracy"])
    sonuclar = sonuclar.append(sonuc, ignore_index=True)

# Doruluk oranlarn grselletirme
sns.barplot(x="Accuracy", y="Modeller", data=sonuclar, color="r")
plt.xlabel("Accuracy %")
plt.title("Modellerin Doruluk Oranlar")
plt.show()

```
