```python
df=pd.read_csv("./diabetes.csv")
y=df["Outcome"]
X=df.drop(['Outcome'],axis=1)
X_train, X_test,y_train, y_test=train_test_split(X,
                                                 y,
                                                 test_size=0.30,
                                                 random_state=42)
```

```python
catb = CatBoostClassifier()
```

```python
catb_params = {
    'iterations': [200, 500, 1000],  # 100 yerine 1000 yazdm
    'learning_rate': [0.01, 0.05, 0.1]
}

# GridSearchCV uygulama
catb_cv_model = GridSearchCV(estimator=catb, param_grid=catb_params, cv=5, n_jobs=-1, verbose=2)
catb_cv_model.fit(X_train, y_train)

# En iyi parametreleri grntleme
print("En iyi parametreler:", catb_cv_model.best_params_)
```
